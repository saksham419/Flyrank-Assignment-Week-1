## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

Pick **one** of these and write a few lines below. This is your Week-1 discovery — you'll reference it in your Week-1 research-question write-up (on the InternHQ board).

**Rules:** describe what you observe as *observed / directional* — never "I proved Google's algorithm." Keep client data out of anything you publish.

```python
df_filtered = df[df["impressions_90d"] > 0]
corr_filtered = df_filtered["search_volume"].corr(df_filtered["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d (for impressions > 0): {corr_filtered:.3f}")

# Compare with the original correlation
# The original correlation 'corr' is available from a previous cell's execution context.
print(f"Original correlation (all pages): {corr:.3f}")

if corr_filtered > corr:
    print("The correlation slightly increases when considering only pages with impressions > 0, but remains very low.")
elif corr_filtered < corr:
    print("The correlation slightly decreases when considering only pages with impressions > 0, but remains very low.")
else:
    print("The correlation remains the same when considering only pages with impressions > 0, and is very low.")
```

```python
ctr_by_content_type_and_pos = visible.groupby(["content_type", "position_tier"])["ctr"].mean().unstack()
print(ctr_by_content_type_and_pos.round(4).to_string())

print("\nObservation: Keyword articles tend to have lower CTRs across all position tiers compared to other content types.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(ctr_by_content_type_and_pos, annot=True, cmap='viridis', fmt=".3f", linewidths=.5)
plt.title('Mean CTR by Content Type and Position Tier')
plt.xlabel('Position Tier')
plt.ylabel('Content Type')
plt.show()
```

### Discovery D — Does content age relate to trend direction?

```python
content_age = df.groupby("trend_direction")["content_age_days"].median()
print(content_age.round(0).to_string())
print("\nObservation: There doesn't appear to be a strong relationship between content age and trend direction based on median values, as the median ages are relatively similar across different trend directions.")
```

### Discovery E — Does average position relate to CTR?

```python
corr_pos_ctr = df["avg_position"].corr(df["ctr"])
print(f"Correlation between avg_position and ctr: {corr_pos_ctr:.3f}")
print("Observation: The negative correlation suggests that as average position increases (meaning the page ranks lower), the CTR tends to decrease. This aligns with the 'CTR cliff' observation.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='avg_position', y='ctr', data=df, alpha=0.5)
plt.title('Scatter Plot of Average Position vs. CTR')
plt.xlabel('Average Position')
plt.ylabel('CTR')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
